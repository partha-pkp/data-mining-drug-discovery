```python
import pandas as pd


df1=pd.read_csv("barreto_smiles.csv")
df2=pd.read_csv("wong_smiles.csv")

df = pd.concat([df1, df2], ignore_index=True)
print(type(df))
print(df.shape)

df_cleaned = df.drop_duplicates(subset=['SMILES'])
print(df_cleaned.shape)


df_cleaned.to_csv('input_data.csv', index=False)
```
